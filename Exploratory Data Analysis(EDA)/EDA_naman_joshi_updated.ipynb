2-D Scatter Plot

```
sns.set_style('darkgrid')
sns.FacetGrid(haberman,hue='status',size=5) \
    .map(plt.scatter,'nodes','age') 

plt.xlabel('nodes')
plt.ylabel('age')
plt.legend()
plt.title('2D-Scatter plot of nodes v/s age')
plt.show()
```

##Observations##
From this scatter plot, we can see that status 1 and 2 are spread across ages and there is not much correlation between the two variables.\
But to name a few \
1) status 1 seems to have lesser nodes when compared to status 2.

Histogram, PDF, CDF

```
sns.set_style('darkgrid')
sns.FacetGrid(haberman,hue='status',size=5) \
    .map(sns.distplot,'age') 
plt.ylabel('probability density')
plt.xlabel('age')
plt.legend()
plt.title('histogram and pdf of age')
plt.show()
```

```
sns.set_style('darkgrid')
sns.FacetGrid(haberman,hue='status',size=5) \
    .map(sns.distplot,'year') 
plt.ylabel('probability density')
plt.xlabel('year')
plt.legend()
plt.title('histogram and pdf of year')
plt.show()
```

```
sns.set_style('darkgrid')
sns.FacetGrid(haberman,hue='status',size=5) \
    .map(sns.distplot,'nodes') 
plt.ylabel('probability density')
plt.xlabel('nodes')
plt.legend()
plt.title('histogram and pdf of nodes')
plt.show()
```

##Observations##
Here we can see that it proves the earlier mention that status 1 has high number of data points equal to 0 nodes

```
haberman_1=haberman[haberman["status"]==1]
haberman_2=haberman[haberman["status"]==2]
```

```
counts, bin_edges=np.histogram(haberman_1['year'], bins=15,density=True)
pdf=counts/(sum(counts))
print(pdf)
print(bin_edges)
cdf=np.cumsum(pdf)
plt.plot(bin_edges[1:],pdf, label='pdf')
plt.plot(bin_edges[1:],cdf,label='cdf')
plt.ylabel('probability')
plt.xlabel('year')
plt.legend()
plt.title('pdf,cdf of year')
```

```
counts, bin_edges=np.histogram(haberman_1['age'], bins=15,density=True)
pdf=counts/(sum(counts))
print(pdf)
print(bin_edges)
cdf=np.cumsum(pdf)
plt.plot(bin_edges[1:],pdf, label='pdf')
plt.plot(bin_edges[1:],cdf,label='cdf')
plt.ylabel('probability')
plt.xlabel('age')
plt.legend()
plt.title('pdf,cdf of age')
```

```
counts, bin_edges=np.histogram(haberman_1['nodes'], bins=15,density=True)
pdf=counts/(sum(counts))
print(pdf)
print(bin_edges)
cdf=np.cumsum(pdf)
plt.plot(bin_edges[1:],pdf, label='pdf')
plt.plot(bin_edges[1:],cdf,label='cdf')
plt.ylabel('probability')
plt.xlabel('nodes')
plt.legend()
plt.title('pdf,cdf of nodes')
```

##Observations##
Nodes having value as 0 has much higher probability which is close to 80%.

```
counts, bin_edges=np.histogram(haberman_1['year'], bins=15,density=True)
pdf=counts/(sum(counts))
print(pdf)
print(bin_edges)
cdf=np.cumsum(pdf)
plt.plot(bin_edges[1:],pdf, label='pdf_year')
plt.plot(bin_edges[1:],cdf,label='cdf_year')

counts, bin_edges=np.histogram(haberman_1['age'], bins=15,density=True)
pdf=counts/(sum(counts))
print(pdf)
print(bin_edges)
cdf=np.cumsum(pdf)
plt.plot(bin_edges[1:],pdf, label='pdf_age')
plt.plot(bin_edges[1:],cdf,label='cdf_age')

counts, bin_edges=np.histogram(haberman_1['nodes'], bins=15,density=True)
pdf=counts/(sum(counts))
print(pdf)
print(bin_edges)
cdf=np.cumsum(pdf)
plt.plot(bin_edges[1:],pdf, label='pdf_nodes')
plt.plot(bin_edges[1:],cdf,label='cdf_nodes')

plt.ylabel('probability')
plt.xlabel('all features')
plt.legend()
plt.title('Comparision of pdf, cdf of all features of haberman_1')
plt.show()
```

```
counts, bin_edges=np.histogram(haberman_2['year'], bins=15,density=True)
pdf=counts/(sum(counts))
print(pdf)
print(bin_edges)
cdf=np.cumsum(pdf)
plt.plot(bin_edges[1:],pdf, label='pdf_year')
plt.plot(bin_edges[1:],cdf,label='cdf_year')

counts, bin_edges=np.histogram(haberman_2['age'], bins=15,density=True)
pdf=counts/(sum(counts))
print(pdf)
print(bin_edges)
cdf=np.cumsum(pdf)
plt.plot(bin_edges[1:],pdf, label='pdf_age')
plt.plot(bin_edges[1:],cdf,label='cdf_age')

counts, bin_edges=np.histogram(haberman_2['nodes'], bins=15,density=True)
pdf=counts/(sum(counts))
print(pdf)
print(bin_edges)
cdf=np.cumsum(pdf)
plt.plot(bin_edges[1:],pdf, label='pdf_nodes')
plt.plot(bin_edges[1:],cdf,label='cdf_nodes')

plt.ylabel('probability')
plt.xlabel('all features')
plt.legend()
plt.title('Comparision of pdf, cdf of all features of haberman_2')
plt.show()
```

##Observations##
haberman_1 has ages starting closer to 30. \
haberman_2 has ages starting in late 30's


Mean, Variance and Std-dev

```
#Mean
print(np.mean(haberman_1['age']))
print(np.mean(haberman_1['nodes']))
print(np.mean(haberman_1['year']))
```

```
print(np.mean(haberman_2['age']))
print(np.mean(haberman_2['nodes']))
print(np.mean(haberman_2['year']))
```

```
#standard deviation
print(np.std(haberman_1['age']))
print(np.std(haberman_1['nodes']))
print(np.std(haberman_1['year']))
```

```
print(np.std(haberman_2['age']))
print(np.std(haberman_2['nodes']))
print(np.std(haberman_2['year']))
```

##Observations##
standard deiation of haberman_2 s higher than haberman_1 by 3.27.

Median, Percentile, Quantile, IQR, MAD

```
#Median
print(np.median(haberman_1['age']))
print(np.median(haberman_1['nodes']))
print(np.median(haberman_1['year']))
```

```
print(np.median(haberman_2['age']))
print(np.median(haberman_2['nodes']))
print(np.median(haberman_2['year']))
```

```
#Quantile
print(np.percentile(haberman_1['age'],np.arange(0,125,25)))
print(np.percentile(haberman_1['year'],np.arange(0,125,25)))
print(np.percentile(haberman_1['nodes'],np.arange(0,125,25)))
```

##Observations##
We can see that more than 50 percentile values of nodes of haberman_1 is 0.

```
print(np.percentile(haberman_2['age'],np.arange(0,125,25)))
print(np.percentile(haberman_2['year'],np.arange(0,125,25)))
print(np.percentile(haberman_2['nodes'],np.arange(0,125,25)))
```

##Observations##
Here the nodes with value 0 is less than 25 percentile.

```
#90 percentile
print(np.percentile(haberman_1['age'],90))
print(np.percentile(haberman_1['year'],90))
print(np.percentile(haberman_1['nodes'],90))
```

```
print(np.percentile(haberman_2['age'],90))
print(np.percentile(haberman_2['year'],90))
print(np.percentile(haberman_2['nodes'],90))
```

##Observations##

```
# Median Absolute Deviation
from statsmodels import robust
print(robust.mad(haberman_1['age']))
print(robust.mad(haberman_1['nodes']))
print(robust.mad(haberman_1['year']))
```

```
print(robust.mad(haberman_2['age']))
print(robust.mad(haberman_2['nodes']))
print(robust.mad(haberman_2['year']))
```
